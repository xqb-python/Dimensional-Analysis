```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics import r2_score

import geatpy as ea
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

##  Import data from Comsol

```python
np.random.seed(0)

#====================================Air-40=====================================
# dataset_path = './空气压差_40组.xlsx'
#====================================Nitrogen-40=====================================
# dataset_path = './氮气压差_40组.xlsx'
#====================================Hydrogen-40=====================================
dataset_path = './hydrogen_P1P4_40.xlsx'
#[微通道宽度，温差和长度]== [H，DeltaT，L]
input_list = ['H', 'DeltaT', 'L','T0','lambda1','lambda2']
#Pressure Rise
output_list = ['Deltap']
df = pd.read_excel(dataset_path)
inputs = df[input_list].to_numpy()
outputs = df[output_list].to_numpy().reshape(-1,)
#尺寸校正
inputs[:,[0,2,4,5]] = 1e-6 * inputs[:,[0,2,4,5]]
#[Lendth,temperature]
dim_matrix = np.array(
    [
        [1., 0., 1.,0,1,1],
        [0., 1., 0.,1,0,0]
    ],
)
```

## Learning dimensionless numbers 

```python
dim = 1
deg = 1
def getcoef(inputs,outputs,GaOpt,deg,dim):
    theta = np.exp(np.log(inputs) @ np.reshape(GaOpt, [inputs.shape[1], dim]))
    poly = PolynomialFeatures(deg)
    X_poly = poly.fit_transform(theta.reshape(-1, 1))
    reg = LinearRegression(fit_intercept=False)
    reg.fit(X_poly, outputs)
    y_pred = reg.predict(X_poly)
    r2 = r2_score(outputs, y_pred)
    print("R2=",r2)
    return reg.coef_
print('****************start find dimensionless*************')
@ea.Problem.single
def evalVars(x):
    #Train a linear model
    theta = np.exp(np.log(inputs) @ np.reshape(x, [inputs.shape[1], dim]))
    poly = PolynomialFeatures(deg)
    X_poly = poly.fit_transform(theta.reshape(-1, 1))
    reg = LinearRegression(fit_intercept=False)
    reg.fit(X_poly, outputs)
    y_pred = reg.predict(X_poly)
    ObjV = r2_score(outputs, y_pred)
    # ObjV = np.max(abs(y_pred-outputs)/outputs)
    
    #constraint
    CV = np.abs((dim_matrix @ x )) #DW=0
    return ObjV,CV

problem = ea.Problem(name='soea quick start demo',
                        M=1,  # Target dimension
                        maxormins=[-1],  # Target Minimize Maximize Tag List, 1: Minimize the target- 1: Maximize this goal
                        Dim = 6,  # Decision variable dimension
                        varTypes=[1]*6,  # List of types of decision variables, 0: real number; 1: Integer
                        lb=[-2]*6,  # low
                        ub=[2]*6,  # up
                        evalVars=evalVars)
# Building Genetic Algorithms
algorithm = ea.soea_SEGA_templet(problem,
                                    ea.Population(Encoding='RI', NIND=400),
                                    MAXGEN=2000,  # Maximum Generation
                                    logTras=0,  # Indicates how many generations log information is recorded, and 0 indicates no logging
                                    trappedValue=1e-6,  # Judgment threshold for stagnation in single objective optimization
                                    maxTrappedCount=80)  # Maximum upper limit value of evolution stagnation counter
# Solve
res = ea.optimize(algorithm, seed=1, verbose=True, 
                  drawing=1, outputMsg=True, drawLog=False, 
                  saveFlag=True, dirName='result')
GaOpt = res['Vars']
```

From the above results, it can be seen that the power of dimensional analysis is [-1,0,0, 0,-1,2], and the corresponding physical quantities are ['H ',' DeltaT ',' L ',' T0 ',' lambda1 ',' lambda2 ']

# Fitting results of different media

```python
import matplotlib.pyplot as plt
GaOpt = [-1,0,0,0,-1,2]
Dimention_y = np.exp(np.log(inputs) @ np.reshape(GaOpt, [6, 1]))
Dimention_x = outputs
plt.figure()
plt.plot(Dimention_x,Dimention_y,'ro')

coef = getcoef(inputs,outputs,GaOpt,deg,dim)
#查看系数的预测能力
theta = np.exp(np.log(inputs) @ np.reshape(GaOpt, [inputs.shape[1], dim]))
if len(coef) == 2:
    y_pred = coef[0] + theta*coef[1]
elif len(coef) == 3:
    y_pred = coef[0] + theta*coef[1] + theta**2 *coef[2] 
# 


print('最大的误差为',np.max(abs(y_pred-outputs)/outputs))

#==========================For Origin software Plot=================================
xplot = np.linspace(0, 0.15,500)
yplot = coef[0] + xplot*coef[1]
plt.figure(figsize=(8,4))
plt.plot(theta,outputs,'ro')
plt.plot(xplot,yplot,'k')
```
